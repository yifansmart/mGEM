```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
CA_data = pd.read_csv("CAvideos.csv")
CA_data.head(10)
```

```python
CA_data = CA_data.drop(['video_id', 'thumbnail_link', 'comments_disabled', 'ratings_disabled', 'video_error_or_removed'], axis = 1)
CA_data.head()
```

```python
CA_data['like_ratio'] = CA_data['likes']/(CA_data['likes'] + CA_data['dislikes'])
CA_data.head()
```

```python
CA_data['comment_view_ratio'] = CA_data['comment_count']/CA_data['views']
CA_data.head()
```

```python
CA_data = CA_data.sort_values('comment_view_ratio', ascending = False)
CA_data.head()
```

```python
CA_data = CA_data.loc[(CA_data['category_id'] == 10) & (CA_data['like_ratio'] >= 0.97)]
CA_data = CA_data.sort_values('views', ascending = False)
CA_data.head()
```
